```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import matplotlib.dates as dts
```

```python
dt = pd.read_csv("https://raw.githubusercontent.com/BearAzul/MiniProject2/seventh/Dataset_UAS/Rainfall_dataset.csv")
dt
```

```python
# Mengatur ukuran gambar (figure) dengan lebar 12 inci dan tinggi 6 inci.
plt.figure(figsize=(10, 5))

# Membuat array x yang berisi urutan angka dari 0 hingga panjang data kolom "Bulan" dalam dt.
x = np.arange(len(dt["Bulan"]))

# Mengubah urutan bulan menjadi dari 1
Bulan = [str(i + 1) for i in range(len(dt["Bulan"]))]

# Menentukan lebar setiap batang dalam diagram batang.
width = 0.28

# Membuat diagram batang dengan menggunakan data dari kolom '1981-1990', '1991-2000', dan '2001-2010'
plt.bar(x + 0.00, dt["1981-1990"], color = "b", width = width, label="1981-1990")
plt.bar(x + 0.28, dt["1991-2000"], color = "r", width = width, label="1991-2000")
plt.bar(x + 0.56, dt["2001-2010"], color = "g", width = width, label="2001-2010")

# Memberikan label sumbu x dan y dengan teks "Bulan" dan "Curah Hujan (mm)".
plt.xlabel("Bulan", fontsize=16, weight="bold")
plt.ylabel("Curah Hujan (mm)", fontsize=16, weight="bold")

# Menampilkan legenda di bagian atas tengah gambar dengan 3 kolom (label untuk setiap periode).
plt.legend(loc="upper center", ncol=3)
plt.xticks(x, Bulan)
plt.grid(zorder=0, color="gray", linestyle="dashed", linewidth=0.5)
plt.show()
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/BearAzul/MiniProject2/seventh/Dataset_UAS/suhu_permukaan.csv")
dataset
```

```python
# Membuat array tanggal dan waktu dari tanggal '2013-07-24' dengan Frekuensi interval setiap 3 jam.
dates = pd.date_range(start='2013-07-24', periods=9, freq='3H')

# Mengatur ukuran gambar (figure) dengan lebar 10 inci dan tinggi 5 inci.
plt.figure(figsize=(10, 5))

# Membuat inisialisasi labels dan color untuk memplot data dari kolom yang sesuai dalam dataset
labels = ['Asimilasi', 'Observasi', 'Non Asimilasi']
colors = ['b', 'y', 'r']

# Membuat loop data dari setiap kolom dalam dataset dengan warna dan label yang sesuai.
for label, color in zip(labels, colors):
  plt.plot(dates, dataset[label], marker='o', color=color, label=label)

# Mengatur tinggi untuk setiap nilai y dalam plot sampai 27
for d in [20, 21, 22, 23, 24, 25, 26, 27]:
  plt.axhline(y=d, color='gray', linestyle='-', linewidth=0.3)

# Mengatur format sumbu x untuk menampilkan jam dan pemosisian utama pada sumbu x secara otomatis
plt.gca().xaxis.set_major_formatter(dts.DateFormatter('%H:%M UTC'))
plt.gca().xaxis.set_major_locator(dts.AutoDateLocator())

# Memberi keterangan pada title dan label
plt.title('Suhu Permukaan', fontsize=16, weight='bold')
plt.xlabel('Waktu', fontsize=14, weight='bold')
plt.ylabel('Celcius', fontsize=14, weight='bold')
plt.legend(loc='lower right')
plt.show()
```
